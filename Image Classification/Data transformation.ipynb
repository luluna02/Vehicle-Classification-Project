```python
size = (244, 244)
resize_transform = transforms.Resize(size)
flipped= transforms.RandomVerticalFlip()
```

```python
for label in os.listdir(DATADIR):
    if label != '.DS_Store':
        path=os.path.join(DATADIR, label)
        for img in os.listdir(path):
            img1 = Image.open(os.path.join(path,img))
            resized_img = resize_transform(img1)
            flipped_img= flipped(resized_img)
            resized_img.save(os.path.join(path,img))
            plt.imshow(flipped_img)
```
